# Performing Sentiment Analysis on raw text
## Steps to perform Textual Preprocessing
### Cleaning the data
1. Converting to lowercase
2. Removing Stopwords and Punctuation

```python
from string import punctuation
from nltk.tokenize import word_tokenize, sent_tokenize
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
nltk.download('stopwords')
```

```python
df = pd.read_csv('amazon.csv/amazon.csv',index_col=False)
df
```

```python
data = df.Text
```

```python
data
```

```python
type(data)
```

```python
text = ""
for i in range(50):
     text += data[i]
```

```python
text
```

```python
# convert this into lowercase 

texts = text.lower()
texts
```

```python
texts = [text for text in texts if text not in punctuation]
texts = ''.join(texts)
```

```python
texts
```

```python
stop_words = set(stopwords.words('english'))
print(stop_words)
```

```python
texts = ''.join(texts)
texts
```

```python
word_token = word_tokenize(texts)
word_token
```

```python
word_token = [word for word in word_token if word not in stop_words]
word_token
```

```python
word_freq = {}
for word in word_token:
    if word in word_freq:
        word_freq[word] += 1
    else:
        word_freq[word] = 1
```

```python
word_freq
```

```python
max_freq = max(word_freq.values())
max_freq
```

```python
prop_word = {}
for key,values in word_freq.items():
    prop_word[key] = values/max_freq
```

```python
prop_word
```

```python
vade = SentimentIntensityAnalyzer()
```

```python
# Get VADER Compound value for sentiment intensity
# tweets['sentiment_intensity'] = [analyzer.polarity_scores(v)['compound'] for v in tweets['edited']]
```

```python
sentiments = [vade.polarity_scores(word)['compound'] for word in word_token]
sentiments
```

```python
positive_words = {}
for word in word_token:
    senti = vade.polarity_scores(word)['compound']
    if senti >= 0.3:
        positive_words[word] = senti
        
```
